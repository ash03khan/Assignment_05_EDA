1. Introduction

This assignment focuses on understanding a dataset using simple analysis and visualization techniques.

The dataset contains information related to fetal health indicators. The purpose of this analysis is to study the data, identify patterns, and understand relationships between different variables.

The goal is not just to analyze the data, but also to explain every step in a simple and understandable way.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Pandas is used for data manipulation and analysis, while NumPy supports numerical computations.
Matplotlib and Seaborn are used for data visualization.

Seaborn is particularly useful for statistical plots such as heatmaps and pair plots,
which help in identifying patterns and relationships in data.

```python
df = pd.read_csv(r"/content/Cardiotocographic.csv")

df.head()
df.info()
```

The dataset is loaded using pandas. The head() function shows the first few rows,
while info() provides details about data types and missing values.

Insight:
Understanding the dataset structure helps identify numerical variables and potential issues
like missing values or incorrect data types.

```python
# Check missing values
df.isnull().sum()

# Fill missing values if any
df.fillna(df.mean(numeric_only=True), inplace=True)
```

Missing values are identified and handled using the mean of the respective columns.

Insight:
Handling missing values ensures that analysis is not affected by incomplete data.
Using mean imputation maintains the overall distribution without removing data.

```python
df.describe()
```

The describe() function provides statistical measures such as mean, median,
standard deviation, minimum, and maximum values.

Insight:
This helps in understanding the central tendency and variability of each feature.
Large differences between min and max values may indicate the presence of outliers.

```python
df.hist(figsize=(14, 10))
plt.tight_layout()
plt.show()
```

Histograms are used to visualize the distribution of numerical variables.

Insight:
Some variables may show normal distribution, while others may be skewed.
Skewed distributions indicate imbalance or extreme values in data.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

Boxplots help in identifying outliers and understanding data spread.

Insight:
Outliers are extreme values that can affect analysis and model performance.
Identifying them helps in deciding whether to remove or treat them.

```python
#Correlation Heatmap (VERY IMPORTANT)
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

A correlation heatmap shows the relationship between variables.

Values close to +1 indicate strong positive correlation,
values close to -1 indicate strong negative correlation,
and values near 0 indicate no correlation.

Insight:
Highly correlated variables may influence each other and can be important for prediction.
This helps in identifying key features affecting fetal health indicators.

```python
sns.pairplot(df)
plt.show()
```

Pair plots show relationships between multiple variables simultaneously.

Insight:
This helps identify patterns, clusters, and linear relationships between variables,
which are useful for deeper analysis.

```python
df.corr()['LB'].sort_values(ascending=False)
```

Correlation with LB (Baseline Fetal Heart Rate) is analyzed.

Insight:
This helps identify which factors influence fetal heart rate.
Variables with higher correlation may have stronger impact on fetal condition.
